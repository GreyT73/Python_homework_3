## Задача 1
### 1.1 
Создать новый признак price_per_sq_lot,
который будет содержать среднюю стоимость за
один кв. метр общей площади


```python
import pandas as pd
houses = pd.read_csv('kc_house_data.csv')
houses['price_per_sq_lot'] = houses['price']/houses['sqft_lot']
houses
```

### 1.2 
Создать новый признак delta_renovated,
который будет содержать разницу в годах между
годом реновацией дома и годом постройки дома.
Если реновации дома не было, то в новом признаке поставьте 0

```python
houses['delta_renovated'] = houses['yr_renovated'] - houses['yr_built']
houses.loc[houses['delta_renovated'] < 0, 'delta_renovated'] = 0
houses.head()

```

### 1.3 
Создайте признаки года продажи, месяца продажи


```python
houses['Год продажи'] = houses['date'].apply(lambda x: x[0:4])
houses['Месяц продажи'] = houses['date'].apply(lambda x: x[4:6])                 
houses.head()
```

### 1.4 
Удалите признаки date, zipcode, lat, long

```python
houses_changed = houses.drop(columns = ['date', 'zipcode', 'lat', 'long']).copy()
houses_changed
```

## Задача 3
### 3.1 
Найдите среднюю стоимость домов в
зависимости от количества спален
Отсортируйте от меньшей стоимости к большей


```python
cost_by_bedrooms = houses.groupby('bedrooms').agg({'price': 'mean'})
cost_by_bedrooms['price'] = cost_by_bedrooms['price'].astype(int)
cost_by_bedrooms.sort_values('price')
```

### 3.2 
Найдите минимальную, среднюю и
максимальную стоимости домов в зависимости от
состояния дома

```python
cost_by_condition = houses.groupby('condition').agg({'price': ['min', 'max','mean']})
cost_by_condition
```

### 3.3 
Постройте таблицу с подсчетом количества
домов в данных в зависимости от вида на
набережную и оценкой вида

```python
pd.crosstab(index=houses['waterfront'], columns=houses['view'])
```

### 3.4 
Каких домов в зависимости от этажности и

### 3.5 
Постройте таблицу с подсчетом медианной
стоимости домов в данных в зависимости от
